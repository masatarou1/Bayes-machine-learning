```python
# 定義式に従って二項分布を計算

import numpy as np
from scipy.special import comb
import matplotlib.pyplot as plt

M = 100
mu = 0.2
Sum_arr = np.zeros(M+1)

for i in range(M+1):
    tmp = comb(M, i) * (mu**i) * ((1-mu) ** (M-i))
    Sum_arr[i] = tmp

plt.plot(range(M+1), Sum_arr)
plt.show()
```

```python
# 定義式に従って多項分布を計算

import math
import numpy as np
from scipy.special import comb
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

M = 5
pi_arr = [0.7, 0.2, 0.1]
Sum_arr = np.zeros([M+1, M+1])

for i in range(M+1):
    for j in range(M+1-i):
        tmp = math.factorial(M) * ((pi_arr[0] ** i) /math.factorial(i)) * ((pi_arr[1] ** j) / math.factorial(j)) * ((pi_arr[2] ** (M-i-j)) / math.factorial(M-i-j))
        Sum_arr[i][j] = tmp

print(Sum_arr)
x = np.arange(0, M+1)
y = np.arange(0, M+1)

X,Y = np.meshgrid(x, y)
Z = Sum_arr

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1)
```

```python
# np.multinomialを使ってカテゴリカル分布を計算

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

N = 5
Cycle_number_for_prob = 100000
Categorical_array_temp = np.zeros((N+1, N+1))
Categorical_array_sum = np.zeros((N+1, N+1))

for k in range(0, Cycle_number_for_prob):
    Categorical_array_temp = np.zeros((N+1, N+1))
    random_categorical = np.random.multinomial(N, [0.7, 0.2, 0.1], 1)
    Categorical_array_temp[random_categorical[0][0], random_categorical[0][1]] +=1
    Categorical_array_sum += Categorical_array_temp

Prob_Categorical_arry_sum = Categorical_array_sum / np.float(N*Cycle_number_for_prob)

print(Prob_Categorical_arry_sum)

x = np.arange(0, N+1)
y = np.arange(0, N+1)

X,Y = np.meshgrid(x, y)
Z = Prob_Categorical_arry_sum

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Prob_Categorical_arry_sum, rstride=1, cstride=1)
```
