```python
import numpy as np
import math
import matplotlib.pyplot as plt

def Cb(a, b):
    return math.gamma(a + b) / (math.gamma(a)* math.gamma(b))

Beta_sum = []

a = 10
b = 40

arr_ab = [[0.5, 0.5], [0.6, 0.8], [1.0, 1.0], [10, 40], [10, 5]]

for j in range(len(arr_ab)):
    Beta_arr = []
    for i in np.arange(0, 1, 0.01):
        Beta_tmp = Cb(arr_ab[j][0], arr_ab[j][1]) * (i ** (arr_ab[j][0] - 1)) * ((1 - i) ** (arr_ab[j][1]-1))
        Beta_arr.append([i, Beta_tmp])
    Beta_sum.append(Beta_arr)


np_Beta_arr = np.array(Beta_sum)


fig, ax = plt.subplots()


for i in range(len(np_Beta_arr)):
    ax.plot(np_Beta_arr[i][:, 0], np_Beta_arr[i][:, 1], label="a = " + str(arr_ab[i][0]) + ":, b = " + str(arr_ab[i][1]))

    ax.legend()



    
plt.show()
```
