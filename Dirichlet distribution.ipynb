```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

alpha = [10, 10, 10]

CD_alpha = math.gamma(alpha[0] + alpha[1] + alpha[2]) / (math.gamma(alpha[0])* math.gamma(alpha[1]) * math.gamma(alpha[2]))
Sum_arr = np.zeros([100, 100])

for i in np.arange(0, 1, 0.01):    
    for j in np.arange(0, 1 - i, 0.01):
        Dirichlet = CD_alpha * (i ** alpha[0]) * (j ** alpha[1]) * ((1 - i - j) ** alpha[2])
        Sum_arr[int(i*100)][int(j*100)] = Dirichlet
        
x = np.arange(0, 1, 0.01)
y = np.arange(0, 1, 0.01)

X,Y = np.meshgrid(x, y)
Z = Sum_arr

fig = plt.figure()
ax = Axes3D(fig)
# ax.plot_surface(X, Y, Z, rstride=1, cstride=1)
ax.scatter(X, Y, Z)
    
```
